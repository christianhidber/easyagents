#### imports

```python
import easyagents
from easyagents.agents import DqnAgent, DuelingDqnAgent
from easyagents.callbacks import plot, log

easyagents.agents.seed = 1
```

### Tf-Agents (Dqn)

```python
%%time

dqnAgent = DqnAgent('CartPole-v0', fc_layers=(100, ), backend='tfagents')
dqnAgent.train([plot.Rewards()], num_iterations=20000, num_iterations_between_eval=1000)
```

### Tensorforce (Dqn)

```python
%%time

dqnAgent = DqnAgent('CartPole-v0', fc_layers=(100, ), backend='tensorforce')
dqnAgent.train([plot.Rewards()], num_iterations=20000, num_iterations_between_eval=1000)
```
